```python
def inverteCaracteres(frase):
    # Retorna um texto com todos os caracteres de uma frase invertido.
    # Entrada: abcd efgh
    # Saída: hgfe dcba
    i = len(frase) - 1
    invertido = ""
    
    while i >= 0:
        invertido = invertido + frase[i]
        i -= 1
        
    return invertido

def newSplit(frase):
    frase = frase + " "
    aux = 0
    listaFrase = []
    auxFrase = ""
    
    while aux < len(frase):
        if frase[aux] != " ":
            auxFrase = auxFrase + frase[aux]
        else:
            listaFrase.append(auxFrase)
            auxFrase = ""
    
        aux += 1
    
    return listaFrase

def inverteCaracteresEmPalavras(frase: str)->str:
    # Retorna um texto com todos os caracteres de cada palavra invertido.
    # Entrada: abcd efgh
    # Saída: dcba hgfe
    frase = newSplit(frase)
    i = 0
    invertido = ""
    
    while i < len(frase):
        j = len(frase[i]) - 1

        if i > 0:
            invertido = invertido + " "
        
        while j >= 0:
            invertido = invertido + frase[i][j]
            j -= 1
        
        i += 1

    return invertido

def invertePalavras(frase: str)->str:
    # Retorna um texto com todas as palavras invertidas.
    # Entrada: abcd efgh
    # Saída: efgh abcd
    frase = newSplit(frase)
    i = len(frase) - 1
    invertido = ""

    while i >= 0:
        invertido = invertido + frase[i]

        if i != 0:
            invertido = invertido + " "
        i -= 1

    return invertido
```
