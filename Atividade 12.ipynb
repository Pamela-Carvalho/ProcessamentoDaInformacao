```python
def matriz_borda_crescente( m: int, n: int ) -> list:
    lista = list()
    i = 0
    while i < m:
        linha = list()
        aux = i    
        for j in range(n):
            linha.append(aux)
            aux += 1
        lista.append(linha)
        i += 1          
    return lista
    # Cria uma matriz em que as diagonais estão listadas em ordem crescente.
    # Exemplo:
    # Entrada: m = 5 e n = 3
    # Saída: [ [0, 1, 2],
    #          [1, 2, 3],
    #          [2, 3, 4],
    #          [3, 4, 5],
    #          [4, 5, 6] ]

def matriz_simetrica(m: int, n: int) -> list:
    matriz = []
    for i in range(m):
        linha = []
        for j in range(n):
            linha.append(abs(i - j))
        matriz.append(linha)
    return matriz
# Cria uma matriz simétrica.
# Exemplo:
# Entrada: m = 4 e n = 4
# Saída: [ [0, 1, 2, 3],
#          [1, 0, 1, 2],
#          [2, 1, 0, 1],
#          [3, 2, 1, 0] ]
    
def eh_matriz_quadrada(matriz: list) -> bool:
    tam_linha = len(matriz[0])
    for i in range(len(matriz)):
        if(tam_linha != len(matriz[i])):
            return False  
    if(tam_linha == len(matriz)):
        return True
    else:
        return False
# Verifica se M é uma matriz (todas as linhas tem a mesma quantidade de
# elementos) e se é quadrada (a quantidade de linha é a mesma que a coluna).

def eh_diagonal(matriz: list) -> bool:
    if(eh_matriz_quadrada(matriz)):
        for i in range(len(matriz)):
            for j in range(len(matriz)):
                if(i != j):
                    if(matriz[i][j] == 0):
                        condicao = True
                    else:
                        return False
    else:
        return False
        
    return condicao
# Verifica se uma matriz é quadrada e se os elementos que não pertencem à
# diagonal principal são iguais a zero.

def matriz_transposta(matriz: list) -> list:
    transposta = list()
    num_linhas = len(matriz)
    num_colunas = len(matriz[0])
    for i in range(num_colunas):
        linha = list()
        for j in range(num_linhas):
            linha.append(matriz[j][i])
        transposta.append(linha)   
    return transposta
# Retorna a matriz transposta de M.
```
